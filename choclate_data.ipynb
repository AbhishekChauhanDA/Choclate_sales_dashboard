```python
total_customer = pd.merge(sales[['order_id','customer_id']],Customers[['gender','customer_id']],on ='customer_id',how='inner')
gender_wise_customer = total_customer.groupby('gender')[['order_id']].nunique().reset_index()
plt.figure(figsize=(6,4))
ax= sns.barplot(data= gender_wise_customer, x='gender',y='order_id')
ax.bar_label(ax.containers[0])
plt.xlabel('')   # remove x label
plt.yticks([])
ax.set_ylabel('')
plt.title('Gender wise total Orders')

plt.show()

```

```python
df = pd.merge(sales, Calander, left_on='order_date', right_on='date', how='left')

monthly_sales = df.groupby(['year','month'])['profit'].sum().reset_index()

monthly_sales['YoY_growth'] = monthly_sales.groupby('month')['profit'].pct_change()*100
```

```python
sns.lineplot(data=monthly_sales, x='month', y='profit',marker= 'o')

plt.title('Year To Date Profit Trend')
plt.show()
```

```python
product_distrition = pd.merge(Product,sales,on='product_id',how='inner')
product_distrition.head()
```

```python
top_products = product_distrition.groupby('product_name')['revenue'] \
    .sum().nlargest(10).index

filtered_df = product_distrition[
    product_distrition['product_name'].isin(top_products)
]

pivot_table = product_distrition.pivot_table(
    index='category',          # Y-axis
    columns='product_name',    # X-axis
    values='revenue',          # Heatmap color
    aggfunc='sum',
    fill_value=0
)
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(pivot_table, cmap='coolwarm', annot=False)

plt.title("Revenue Heatmap (Category vs Product)")
plt.xlabel("Product Name")
plt.ylabel("Category")

plt.show()
```

```python
bins = [0, 18, 30, 40, 50, float('inf')]
labels = ['Under 18', '18-30', '31-40', '41-50', 'Above 50']

Customers['age_group'] = pd.cut(
    Customers['age'],
    bins=bins,
    labels=labels,
    right=False
)
Customers.head()
```

```python
age_wise_sales= pd.merge(Customers,sales,on='customer_id',how='inner')
agg_data = (
    age_wise_sales
    .groupby(['age_group', 'gender'],observed=True)['revenue']
    .sum()
    .reset_index()
)
plt.figure(figsize=(12,6))
ax=sns.barplot(data=age_wise_sales,x='age_group',y='revenue',hue='gender')
plt.title('Age group wise total sales')
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```

```python
coutnry_sale= pd.merge(store,sales,on='store_id',how='inner')
total_sale = coutnry_sale.groupby('country')[['revenue','profit']].sum().reset_index()
total_sale['profit_margin_%'] = (
    total_sale['profit'] / total_sale['revenue']
) * 100
total_sale.head(10)
```

```python
plt.figure(figsize=(12,6))

ax = sns.barplot(
    data=total_sale,
    x='country',
    y='revenue'
)
ax.bar_label(ax.containers[0])

plt.title('Country-wise Total Sales')

plt.show()
```

```python
total_unit = pd.merge(store,sales,on='store_id',how="inner")
total_unit = total_unit.groupby('city')['order_id'].count().reset_index()
plt.figure(figsize=(12,6))

ax = sns.barplot(
    data=total_unit,
    x='city',
    y='order_id'
)
ax.bar_label(ax.containers[0])
plt.title("city wise total slaes")
plt.ylabel(" ")
plt.xlabel(" ")
plt.yticks([])        
plt.show()
```

```python
gender_sales= pd.merge(Customers,sales,on='customer_id',how='inner')
gender_sales = gender_sales.groupby('gender')['revenue'].sum().reset_index()
plt.figure(figsize=(6,6))

plt.pie(
    gender_sales['revenue'],
    labels=gender_sales['gender'],
    autopct='%1.1f%%',
    startangle=90,
    wedgeprops={'width':0.4}   # 👈 makes it donut
)

plt.title('Gender-wise Total Sales (Donut Chart)')

plt.show()
```
